## Sentiment Analysis 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('overallProduct.csv')
df1 = df[['name','reviewstext','reviewsrating']]
df1.groupby('name').count()
```

```python
df1.groupby('name').mean().sort_values(["reviewsrating"],ascending=False)
```

```python
#convert text to string
for i in range(0,len(df1)-1):
    if type(df1.iloc[i]['reviewstext']) != str:
        df1.iloc[i]['reviewstext'] = str(df1.iloc[i]['reviewstext'])
```

```python
#Drop rating = 3
amazon = df1[df1['reviewsrating'] != 3]
```

```python
#Assign 1 and 0 for rating >= 4 and <= 2
def sentiment(n):
    if n >= 4:
        return 1
    else: return 0
amazon['sentiment'] = amazon.loc[:,'reviewsrating'].apply(sentiment)
amazon.head()
```

```python
def alltogeter (row):
    return row['name'] + ' '+ row['reviewstext']
amazon['name_text'] = amazon.apply(alltogeter, axis=1)
amazon.head(50)
```

```python
X = amazon['name_text']
y = amazon['sentiment']
```

```python
#75% Train and 25%test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
workcount = CountVectorizer()
numTFTR = workcount.fit_transform(X_train)
numTFtest = workcount.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(numTFTR, y_train)
```

```python
predict = model.predict(numTFtest)
```

```python
for i in predict:
    print(i, end=" ")
```

```python
predicts = pd.DataFrame(predict)
predicts.to_csv("sentiment.csv",index = False)
```

```python
accuracy_score(y_test, predict)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('overallProduct.csv')
df2 = df[['name','reviewstext','reviewsrating']]
df2
```

```python
df2[df2['name']=='Amazon Echo Show Alexa-enabled Bluetooth Speaker with 7" Screen']
df2.dtypes
```

```python
df2[['reviewsrating']] = df2[['reviewsrating']].astype(str)
df2.dtypes
```

```python
array = ['4','5']
df2 = df2.loc[(df2['name'] == 'Amazon Echo Show Alexa-enabled Bluetooth Speaker with 7" Screen') & df2['reviewsrating'].isin(array)]
```

```python
df2['reviewstext']= df2['reviewstext'].str.lower()
df2
```

```python
df2['reviewstext']= df2['reviewstext'].str.replace('\W+', ' ',regex =True)
df3 = df2[['reviewstext']]
```

```python
df3['reviewstext']= df3['reviewstext'].str.replace('\d+', '',regex =True)
```

```python
dicto = {'äôt':'','äôve':'','äù':'','äú':'','äúalexa':'','äúbest':'','äúdropping':'','äúshow':'','äústar':'','äúthings':'','äôm':'','äôs':'',
         'äî':'','äîbadly':'','äîespecially':'','äò':'','äôd':'','äôll':'','äôre':''}
```

```python
df3['reviewstext']= df3['reviewstext'].replace(dicto,regex = True)
df3
```

```python
vectorizer = CountVectorizer()
matrix = vectorizer.fit_transform(df3['reviewstext'])
matrix 
```

```python
matrix.toarray()
```

```python
print(vectorizer.get_feature_names())
```

```python
counts = pd.DataFrame(matrix.toarray(),columns=vectorizer.get_feature_names())
s = counts.sum(axis = 0)
```

```python
s = s.sort_values(ascending=False)
s.count()
```
