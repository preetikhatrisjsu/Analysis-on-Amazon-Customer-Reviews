```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('filled_Product.csv')
```

```python
df.index = np.arange(1, len(df)+1)
```

```python
df.head()
```

```python
df['dateAdded'] = df['dateAdded'].str.extract(r'^(\s*\d+\s*\W+\s*\d+\s*\W+\s*\d+)\S*')
```

```python
df['reviewsdate'] = df['reviewsdate'].str.extract(r'^(\s*\d+\s*\W+\s*\d+\s*\W+\s*\d+)\S*')
```

```python
df.insert(loc=0,column='ProductID',value=df['productCode'].factorize()[0]+1)
```

```python
df.insert(loc=13,column='usernameID',value=df['reviewsusername'].factorize()[0]+1)
```
